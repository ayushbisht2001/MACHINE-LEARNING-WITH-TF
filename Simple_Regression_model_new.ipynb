```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

**numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)[source]**

*Returns num evenly spaced samples, calculated over the interval [start, stop]*

```

x_data = np.linspace(0,10,50) + np.random.uniform(-1.5,1.5,50)
print(x_data)
```

```
y_label = np.linspace(0,10,50)  + np.random.uniform(-1.5,1.5,50)
```

```
plt.plot(x_data,y_label, marker='o', linestyle='dashed', linewidth=1, markersize=5)
```

**y = mx +b**

follow this link for eager_execution
https://www.tensorflow.org/guide/eager


```
m = tf.Variable(0.98)   # these are tensor's object and behave as a node
b = tf.Variable(0.85)

# The zip() function takes iterables (can be zero or more), aggregates them in a tuple, and return it. 
print(set(zip(x_data,y_label)))

# tf.compat.v1.disable_eager_execution()
```

```
for x,y in zip(x_data,y_label):
  y_hat = m*x + b
  error = tf.square((y-y_hat))
```

**tf.compat.v1.train.GradientDescentOptimizer()**

The major line of interest here is train_op = tf.train.GradientDescentOptimizer(0.01).minimize(error) where the training step is defined. It aims to minimise the value of the errorVariable, which is defined earlier as the square of the differences (a common error function). The 0.01 is the step it takes to try learn a better value.

```
# tf.compat.v1.disable_eager_execution()
optimizer = tf.compat.v1.train.GradientDescentOptimizer(learning_rate= 0.001)
train = optimizer.minimize(error)
```

*for deep go to https://cs230.stanford.edu/blog/moretensorflow/*

here, we have initialised global variable , with this thing , we can access the variable or access its value .

there is another way to access the variables value , chekout out from the above link.




```
init = tf.compat.v1.global_variables_initializer()
```

# Session

It is used to access the tensor flow object like tf.constant() or tf.Variable() ... etc

we can not directly access them.

```
with tf.compat.v1.Session() as sess:
  sess.run(init)  # initialising the variables
  epoch = 1000
  for i in range(epoch):
    sess.run(train)
    print(" m =  ",sess.run(m) ," ---- ", "b = ",sess.run(b))
    print("error" , sess.run(error))
    
  final_slope , final_intercept = sess.run([m,b])
    
```

```
final_slope
```

```
x_test = np.linspace(0,10,50)
y_pred_plot = final_slope * x_test + final_intercept
plt.plot(x_test,y_pred_plot,marker='+',markersize=4,linestyle="dashed",linewidth=1)
plt.plot(x_data,y_label,'*')
```
