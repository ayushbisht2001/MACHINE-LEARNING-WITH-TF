```
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
from google.colab import files
import io

uploaded = files.upload()
```

```
data = pd.read_csv(io.BytesIO(uploaded["heart.csv"]))
data = data.sample(frac = 1)
no_rows = data.size//data.columns.size

data.index = range(0,no_rows)
```

```
data.head(5)
```

```
 col_to_norm = ["trestbps", "chol", "age", "thalach", "oldpeak", "slope", "thal"]
```

```
data[col_to_norm]
```

```
data[col_to_norm] = data[col_to_norm].apply(lambda x: (x - x.min())/(x.max() - x.min()) )
```

```
data[col_to_norm].head(5)
```

```
feat_trestbps = tf.feature_column.numeric_column("trestbps")
feat_chol = tf.feature_column.numeric_column("chol")
feat_age = tf.feature_column.numeric_column("age")
feat_thalach = tf.feature_column.numeric_column("thalach")
feat_oldpeak = tf.feature_column.numeric_column("oldpeak")
feat_slope = tf.feature_column.numeric_column("slope")
feat_thal = tf.feature_column.numeric_column("thal")
```

```
from tensorflow.keras import layers
```

```
feat_cols = [feat_trestbps, feat_chol, feat_age, feat_thalach, feat_oldpeak, feat_slope, feat_thal]
```

```
x_data = data.drop("target",axis = 1)
```

```
x_data.head(4)
```

```
x_data.info()
```

```
labels = data["target"]
```

```
labels.head()
```

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(x_data,labels,test_size = 0.3,random_state = 102)
```

```
x_train
```

```
inp_func = tf.compat.v1.estimator.inputs.pandas_input_fn(x= x_train, y=y_train, batch_size=10, num_epochs = 100, shuffle = True )
```

```
model = tf.compat.v1.estimator.LinearClassifier(feature_columns=feat_cols, n_classes=2)
```

```
model.train(input_fn=inp_func, steps=100)
```

```
evaluate_inp_fun = tf.compat.v1.estimator.inputs.pandas_input_fn(x=x_test, y=y_test, num_epochs = 100, batch_size = 10, shuffle = True)
```

```
result = model.evaluate(evaluate_inp_fun)
```

```
result
```

```
predict_inf_fun = tf.compat.v1.estimator.inputs.pandas_input_fn(x= x_test, y=y_test, num_epochs=1, batch_size=10, shuffle = False)
```

```
predictions = model.predict(predict_inf_fun)
```
